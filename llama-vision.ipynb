```python
ocr=OCRProcessor(model_name='llama3.2-vision:11b')
for file in os.listdir('../../data/wee-kim-wee/scan/combined/'):
    if not file.startswith('.'):
        image_path = os.path.join('../../data/wee-kim-wee/scan/combined/', file)
        ocr_path = os.path.join('../../data/wee-kim-wee/pytesseract/combined/', file.replace('.png','')+'_pytesseract'+'.txt')
        with open(ocr_path, 'r') as f:
            pytesseract_ocr = f.read()
            f.close()
        text = ocr.process_image(image_path,
            format_type='text',
            custom_prompt='"text": f"""Extract all visible text from this image in English **without any changes**  \\ \
                                - **Do not summarize, paraphrase, or infer missing text.** \\ \
                                - Retain all spacing, punctuation, and formatting exactly as in the image. \\ \
                                - If text is unclear or partially visible, extract as much as possible without guessing. \\ \
                                - **Include all text, even if it seems irrelevant or repeated.**"""  \\ \
                                Correct the OCR errors with the scanned image. Text:' + pytesseract_ocr)
        with open('../../data/wee-kim-wee/llama-vision/combined/'+file.replace('.png','')+'_llama-vision'+'.txt', 'w') as f:
            f.write(text)
            f.close()
            print('Processed:', file)
```

```python
from ollama_ocr import OCRProcessor

# Initialize OCR processor
ocr = OCRProcessor(model_name='llama3.2-vision:11b', max_workers=4)
# Batch process images
batch_results = ocr.process_batch(
    input_path="path/to/images/folder",
    format_type="json",
    recursive=True,      # Include subdirectories
    preprocess=True      # Apply image preprocessing
)
# Iterate through results
for file_path, text in batch_results['results'].items():
    print(f"\nFile: {file_path}")
    print(f"Extracted Text: {text}")
# Display statistics
print(f"\nTotal images: {batch_results['statistics']['total']}")
print(f"Successfully processed: {batch_results['statistics']['successful']}")
print(f"Failed: {batch_results['statistics']['failed']}")
```
